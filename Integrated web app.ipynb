```python
import tensorflow as tf
from keras.models import load_model
import numpy as np
from PIL import Image
import gradio as gr

# Load the saved model
model = load_model('emergency_classifier.h5')

# Define the prediction function
def predict_image(image):
    # Preprocess the image

    print(image[1])
    
  
    
    image = np.expand_dims(image, axis=0)
    

    # Make predictions
    result = model.predict(image)
    if result[0][0] == 1:
        prediction = 'Non-emergency vehicle'
    else:
        prediction = 'Emergency vehicle'

    return prediction

# Create the interface
iface = gr.Interface(fn=predict_image, 
                      inputs=gr.inputs.Image(label="Upload Image", image_mode='RGB', shape=(64, 64)),
                     outputs=gr.outputs.Textbox(),
                     title="Emergency Vehicle Classifier",
                     description="Upload an image to classify if it's an emergency vehicle or not.")

# Launch the interface
iface.launch()
```
